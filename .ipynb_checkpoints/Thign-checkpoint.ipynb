```python
import numpy as np
from scipy.interpolate import UnivariateSpline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

def graphFrequency(location, EddysplinePhase, EddysplineGain, ChargesplinePhase, ChargesplineGain,splineEddyGainDisp):
    ActualData = pd.read_csv(location)
    ActualData = ActualData.groupby("Frequency").mean()
    
    ActualData["LogFreq"] = np.log10(ActualData.index)
    
    
    ActualData["Displacement (micron)"] = splineEddyGainDisp(ActualData["Amplitude B"])
    
    
  
    
    ActualData["ChargeGain"] = ChargesplineGain(ActualData["LogFreq"])
    ActualData["ChargePhase"] = ChargesplinePhase(ActualData["LogFreq"])


    ActualData["EddyGain"] = EddysplineGain(ActualData["LogFreq"])
    ActualData["EddyPhase"] = EddysplinePhase(ActualData["LogFreq"])

    ActualData["totalPhase"] = ActualData["EddyPhase"] + ActualData["ChargePhase"]
    ActualData["totalGain"] = ActualData["EddyGain"] * ActualData["ChargeGain"]
    
    
    ActualData["Fixed Phase"] = ((ActualData["deltaPhaseAngle"]-ActualData["totalPhase"]))
    ActualData["Fixed Amplitude"] = ActualData["Amplitude A"]/ActualData["ChargeGain"]
    
    ActualData["Fake Modulus"] = ActualData["Fixed Amplitude"] /ActualData["Displacement (micron)"]
    ActualData["Fake G'"] = ActualData["Fake Modulus"] * np.cos(3.14/180*ActualData["Fixed Phase"] )
    ActualData["Fake G''"] = ActualData["Fake Modulus"] * np.sin(3.14/180*ActualData["Fixed Phase"] )
    
    print(location)
    

    plt.figure()

    plt.plot(ActualData["LogFreq"],ActualData["Fixed Phase"], 'bo', ms=5) # original data in red

    plt.show()
    
    
    plt.figure()

    plt.plot(ActualData["LogFreq"],ActualData["Fixed Amplitude"], 'go', ms=5) # original data in red

    plt.show()
    
    plt.figure()

    plt.plot(ActualData["LogFreq"],ActualData["Fake G'"], 'ro', ms=5, label = "G'") # original data in red
    plt.plot(ActualData["LogFreq"],ActualData["Fake G''"], 'go', ms=5, label = "G''") # original data in red
    
    plt.yscale("log")
    plt.legend()
    plt.show()
    return ActualData

########################













NewPDMS2 = "C:/Users/khperera/Documents/GraphOutput/pd,sconnected_improvesignalOut__channelA__test39_parallel_fullcleanred_groundeddriber.csv"
pdms2 = graphFrequency(NewPDMS2,splineEddyPhase, splineEddyGain, splineChargePhase, splineChargeGain,splineEddyGainDisp)



NewSilicon2 = "C:/Users/khperera/Documents/GraphOutput/siliconoilconnected_improvesignalOut__channelA__test39_parallel_fullcleanred_groundeddriber.csv"
NewSilicon = graphFrequency(NewSilicon2,splineEddyPhase, splineEddyGain, splineChargePhase, splineChargeGain,splineEddyGainDisp)

NewSilicon["Difference"] =  (90-pdms2["deltaPhaseAngle"] + NewSilicon["deltaPhaseAngle"])+3
NewSilicon["Difference2"] =  (180+pdms2["Fixed Phase"] - NewSilicon["Fixed Phase"])%180
plt.plot(NewSilicon["LogFreq"],NewSilicon["Difference"], 'bo', ms=5)
plt.show()
NewSilicon["Fake Modulus2"] = NewSilicon["Fake Modulus"]/pdms2["Fake Modulus"]
NewSilicon["Fake G'"] = NewSilicon["Fake Modulus"] * np.cos(NewSilicon["Difference"]*3.14/180 )
NewSilicon["Fake G''"] = NewSilicon["Fake Modulus"] * np.sin(NewSilicon["Difference"]*3.14/180  )
NewSilicon["Fake G'2"] = NewSilicon["Fake Modulus2"] * np.cos(NewSilicon["Difference"]*3.14/180 )
NewSilicon["Fake G''2"] = NewSilicon["Fake Modulus2"] * np.sin(NewSilicon["Difference"]*3.14/180  )
plt.plot(NewSilicon["LogFreq"],NewSilicon["Fake G'"], 'bo', ms=5, label = "G'")
plt.plot(NewSilicon["LogFreq"],NewSilicon["Fake G''"], 'go', ms=5, label = "G''")
plt.yscale("log")
plt.legend()
plt.show()
plt.plot(NewSilicon["LogFreq"],NewSilicon["Fake G'2"], 'bo', ms=5, label = "G'")
plt.plot(NewSilicon["LogFreq"],NewSilicon["Fake G''2"], 'go', ms=5, label = "G''")
plt.yscale("log")
plt.legend()
plt.show()
#plt.plot(NewSilicon["LogFreq"],NewSilicon["Difference2"], 'ro', ms=5)

#Splt.ylim(-10,10)
# original data in red
#plt.plot(siliOil2["LogFreq"],siliOil2["G''"], 'go', ms=5) # original data in red
#plt.yscale("log")
```
